```python
import pygame

#Colors
WHITE = (255, 255, 255)
BLUE = (0, 0, 255)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLACK = (0, 0, 0)
pygame.init()
# Initializing pygame

size = (800, 600)
screen = pygame.display.set_mode(size)
pygame.display.set_caption("pong game")

# Adding Music 
    
from pygame import mixer
mixer.init()

mixer.music.load("computerlove.mp3")

mixer.music.set_volume(0.6)

mixer.music.play()

# Starting coordinates of the paddle
baseRectangle_x = 400
baseRectangle_y = 580

# initial speed of the paddle
base_r_X = 0
base_r_Y = 0

# beginning position of the ball
ball_x = 50
ball_y = 50

# speed of the ball
ball_change_x = 5
ball_change_y = 5
score = 0

# draws the paddle and restricts movement between the window
def drawrect(screen, x, y):
    if x <= 0:
        x = 0
    if x >= 699:
        x = 699
    pygame.draw.rect(screen, WHITE, [x, y, 100, 20])
    
#main game loop

active = False
clock = pygame.time.Clock()
while not active:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            active = True
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                base_r_X = -6
            elif event.key == pygame.K_RIGHT:
                base_r_X = 6
        elif event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                base_r_X = 0
            elif event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                base_r_Y = 0
    screen.fill(BLACK)
    baseRectangle_x += base_r_X
    baseRectangle_y += base_r_Y
    ball_x += ball_change_x
    ball_y += ball_change_y
    
    # Controls movement of the ball
    if ball_x < 0:
        ball_x = 0
        ball_change_x = ball_change_x * -1
    elif ball_x > 785:
        ball_x = 785
        ball_change_x = ball_change_x * -1
    elif ball_y < 0:
        ball_y = 0
        ball_change_y = ball_change_y * -1
    elif ball_x > baseRectangle_x and ball_x < baseRectangle_x + 100 and ball_y == 565:
        ball_change_y = ball_change_y * -1
        score = score + 1
    elif ball_y > 600:
        ball_change_y = ball_change_y * -1
        score = 0
        
    #Level control
    if score <5:
        level = 1
    if score >5 and score<10:
        level = 2
    if score >10 and score<20:
        level = 3
    if score >20 and score<30:
        level=4
    if score >30 and score<40:
        level=5
    if score >40 and score<50:
        level=6
    if score>60 and score<70:
        level=7
    if score>70 and score<80:
        level=8
    if score>80 and score<90:
        level=9
    if score>90 and score<100:
        level=10
        
    pygame.draw.rect(screen, RED, [ball_x, ball_y, 15, 15])
    
    # Draws the ball
    drawrect(screen, baseRectangle_x, baseRectangle_y)
    
    # Score Board
    font = pygame.font.SysFont('freesansbold.ttf', 25, False, False)
    text = font.render("Score = " + str(score), True, BLUE)
    screen.blit(text, [360, 10])
    
    #Level board
    font1=pygame.font.SysFont('freesansbold.ttf', 25, False, False)
    text1=font.render("Level = "+ str(level), True, GREEN)
    screen.blit(text1, [10,10])
    pygame.display.flip()

    clock.tick(110)
    
    
pygame.quit()
```
